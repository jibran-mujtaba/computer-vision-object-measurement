```python
import cv2

img = cv2.imread(r"D:\Desktop\XIS-Assesment\computer-vision-object-measurement\dataset\test\images\1000141687_jpg.rf.4a3a2274e6286be3c8148e850aa366402.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(gray)

for cols in range(4, 11):
    for rows in range(4, 11):
        found, _ = cv2.findChessboardCornersSB(
            gray, (cols, rows),
            flags=cv2.CALIB_CB_NORMALIZE_IMAGE | cv2.CALIB_CB_EXHAUSTIVE
        )
        if found:
            print(f"MATCH: inner corners = ({cols}, {rows})")
```
